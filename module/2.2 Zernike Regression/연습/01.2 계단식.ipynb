```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial

# Zernike radial polynomial
def zernike_radial(n, m, r):
    radial = np.zeros_like(r)
    for s in range((n - m) // 2 + 1):
        c = (-1)**s * factorial(n - s) / (
            factorial(s) * factorial((n + m) // 2 - s) * factorial((n - m) // 2 - s)
        )
        radial += c * r**(n - 2 * s)
    return radial

# Zernike polynomial
def zernike_polynomial(n, m, r, theta):
    R = zernike_radial(n, abs(m), r)
    if m >= 0:
        return R * np.cos(m * theta)
    else:
        return R * np.sin(abs(m) * theta)

# 계단형 Zernike 다항식 플롯
def plot_zernike_staircase(max_n):
    fig, ax = plt.subplots(figsize=(20, 15))
    dx, dy = 2, 2  # 플롯 간격

    # 모든 \(n, m\)에 대해 Zernike 다항식 플롯
    for n in range(max_n + 1):  # n은 m + k
        for m in range(-n, n + 1, 2):  # m은 -n ~ n (짝수만)
            # Create polar coordinates
            r = np.linspace(0, 1, 500)
            theta = np.linspace(0, 2 * np.pi, 500)
            r, theta = np.meshgrid(r, theta)
            x = r * np.cos(theta)
            y = r * np.sin(theta)
            z = zernike_polynomial(n, m, r, theta)

            # 좌표 계산 (계단형 배치)
            x_offset = m * dx
            y_offset = -n * dy

            # Zernike 다항식 플롯
            ax.pcolormesh(x + x_offset, y + y_offset, z, shading='auto', cmap='rainbow')
            ax.text(
                x_offset,
                y_offset - 0.8,  # 제목 위치
                f"$Z_{{{n}}}^{{{m}}}$",
                color="black",
                fontsize=15,
                ha="center",
            )

    # 플롯 설정
    ax.set_aspect('equal')
    ax.axis('off')  # 축 숨기기
    plt.title("Zernike Staircase Plot", fontsize=16)
    plt.show()

# 실행: 최대 \(n\) 값을 설정하여 플롯 생성
plot_zernike_staircase(max_n=6)

```
