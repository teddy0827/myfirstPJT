```python
import numpy as np
import matplotlib.pyplot as plt

def visualize_zernike(n, m):
    # 극좌표 설정
    r = np.linspace(0, 1, 500)
    theta = np.linspace(0, 2 * np.pi, 500)
    R, Theta = np.meshgrid(r, theta)
    
    # Zernike 다항식 계산 (예시)
    if m >= 0:
        Z = R**n * np.cos(m * Theta)  # 코사인 성분
    else:
        Z = R**n * np.sin(-m * Theta)  # 사인 성분
    
    # 데카르트 좌표로 변환
    X = R * np.cos(Theta)
    Y = R * np.sin(Theta)
    
    # 플롯
    plt.figure(figsize=(4, 4))
    plt.contourf(X, Y, Z, levels=100, cmap='rainbow')  # cmap='rainbow' 사용
    plt.colorbar(label="Amplitude")
    plt.title(f"Zernike Polynomial (n={n}, m={m})")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.axis('equal')
    plt.show()

# 테스트 실행 (n=3, m=2)
visualize_zernike(0, 0)
visualize_zernike(1, 1)
visualize_zernike(3, 1)
```

n=3, m=1 대변못함


```python
import numpy as np
import matplotlib.pyplot as plt

def visualize_multiple_zernike(max_n):
    # 필요한 subplot 개수 계산
    total_plots = sum([2 * n + 1 for n in range(max_n + 1)])
    fig, axes = plt.subplots(max_n + 1, max_n * 2 + 1, figsize=(15, 15), 
                             gridspec_kw={'wspace': 0.5, 'hspace': 0.5})

    for n in range(max_n + 1):
        for m in range(-n, n + 1):
            # 인덱스 계산
            row = n
            col = m + max_n  # 가운데 정렬을 위한 오프셋
            ax = axes[row, col]

            # Zernike 다항식 계산
            r = np.linspace(0, 1, 500)
            theta = np.linspace(0, 2 * np.pi, 500)
            R, Theta = np.meshgrid(r, theta)

            if m >= 0:
                Z = R**n * np.cos(m * Theta)
            else:
                Z = R**n * np.sin(-m * Theta)

            X = R * np.cos(Theta)
            Y = R * np.sin(Theta)

            # 서브플롯 생성
            c = ax.contourf(X, Y, Z, levels=100, cmap='rainbow')
            ax.set_title(f"$Z_{{{n}}}^{{{m}}}$", fontsize=8)
            ax.axis('equal')
            ax.axis('off')

    # 컬러바 추가
    fig.colorbar(c, ax=axes[:, -1], shrink=0.5, label="Amplitude")
    plt.tight_layout()
    plt.show()

# 테스트 실행 (예: 최대 5차)
visualize_multiple_zernike(5)

```
