```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#   .
file_path = 'C:/vscode/MODULE/modules2/08.  /output.xlsx'  #    
df = pd.read_excel(file_path, sheet_name='Ideal TROCS')

#    
columns_to_analyze = [
    'X_reg_demrc', 'Y_reg_demrc', 'raw_x', 'raw_y', 'pred_x', 'pred_y', 'residual_x', 'residual_y', 
    'cpe_pred_x', 'cpe_pred_y', 'cpe_resi_x', 'cpe_resi_y', 'mrc_fit_x', 'mrc_fit_y',
    'trocs_fit_x', 'trocs_fit_y', 'residual_x_detrocs', 'residual_y_detrocs', 
    'ideal_trocs_x', 'ideal_trocs_y'
]

# STEPSEQ   .
unique_stepseq = df['M_STEP'].unique()
colors = plt.cm.get_cmap('Set1', len(unique_stepseq))

#      (4x5 ,     )
n_rows, n_cols = 5, 4
fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, 20))
axes = axes.ravel()  #   axes 1D array   

#      .
for i, column in enumerate(columns_to_analyze):
    grouped_data = df.groupby('UNIQUE_ID').agg(
        mean_col=(column, 'mean'),
        std_col=(column, 'std'),
        P_TIME=('P_TIME', 'first'),
        M_STEP=('M_STEP', 'first')
    )
    grouped_data['mean_plus_3sigma'] = grouped_data['mean_col'] + 3 * grouped_data['std_col']

    ax = axes[i]  #  subplot 
    for j, m_step in enumerate(unique_stepseq):
        subset = grouped_data[grouped_data['M_STEP'] == m_step]
        ax.scatter(pd.to_datetime(subset['P_TIME']), subset['mean_plus_3sigma'], 
                   color=colors(j), label=f'M_STEP: {m_step}', alpha=0.6)
    
    ax.set_title(f'{column} (mean + 3 sigma)')
    ax.set_xlabel('P_TIME')
    ax.set_ylabel(f'{column}')
    ax.grid(True)

#    ,   
plt.tight_layout()
plt.legend(title='M_STEP', loc='upper right', bbox_to_anchor=(1.2, 1))
plt.show()
```
